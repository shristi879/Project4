```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# Confirm that our new database was created
print(mongo.list_database_names())
```

```python
# Assign the uk_food database to a variable name
db = mongo['hrdata']
```

```python
# Review the collection in our new database
print(db.list_collection_names())
```

```python
# review a document in the establishments collection
pprint(db.employee.find_one())
```

```python
employee = db['employee']
```

```python
query = {'Attrition': 'No'}
results = employee.find(query)
for result in results:
    pprint(result)
```

```python
query = {}
sort = [('Business Travel', 1)]

results = employee.find(query).sort(sort)

for i in range(5):
    pprint(results[i])
```

```python
query = {}
sort = [('Business Travel', 1)]
limit = 5

pprint(list(employee.find(query).sort(sort).limit(limit)))
```
